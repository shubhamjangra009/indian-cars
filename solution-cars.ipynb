## 1.1. Understand Dataset:

```python
cars = pd.read_csv('cars_ds_final.csv')
cars = cars.iloc[:, 1:]
cars
```

## 1.2. Data Cleaning: 
___Handle missing values, if any, by either removing or imputing them based on the nature of the data.___

Get a sense of how many empty values are there

```python
pd.DataFrame({'No. of Empty Values':cars.isna().sum()})
```

```python
condition = cars['Make'].isna()
cars[condition][['Make', 'Model']]
```

```python
cars[condition]['Model'].value_counts()
```

Replacing all 'Make' column values with appropriate brand name

```python
cars.loc[cars.Make.isna() & cars.Model.str.startswith('Mercedes'), 'Make'] = 'Mercedes'
cars.loc[cars.Make.isna() & cars.Model.str.startswith('Go+'), 'Make'] = 'Datsun'
cars.loc[cars.Make.isna() & cars.Model.str.startswith('Rolls'), 'Make'] = 'Rolls-Royce'
```

Excluding null values of Displacement column

```python
cars = cars[~cars.Displacement.isna()]
```

Excluding null values of Cylinders column

```python
cars = cars.dropna(subset='Cylinders')
```

```python
cars.Valves_Per_Cylinder.unique(), cars.Valves_Per_Cylinder.value_counts()
```

Imputing null values with most frequent value _i.e. 4_ for column Valves_Per_Cylinder

```python
cars.Valves_Per_Cylinder.fillna(cars.Valves_Per_Cylinder.mode().values[0], inplace=True)
```

```python
cars.Drivetrain.value_counts()
```

```python
cars.loc[cars.Drivetrain == '4WD', 'Drivetrain'] = '4WD (Four Wheel Drive)'
```

```python
cars[cars.Drivetrain.isnull()]
```

Imputing Drivetrain column with appropriate drivetrain value according to car

```python
cars.loc[cars.Drivetrain.isnull() & cars.Make.isin(['Mercedes']), 'Drivetrain'] = 'RWD (Rear Wheel Drive)'
cars.loc[cars.Drivetrain.isnull(), 'Drivetrain'] = 'FWD (Front Wheel Drive)'
```

```python
cars.isna().sum()
```

Excluding null values of Cylinder_Configuration, Emission_Norm, Engine_Location column

```python
cars.dropna(subset=['Cylinder_Configuration', 'Emission_Norm', 'Engine_Location'], inplace=True)
```

```python
cars.loc[cars.Emission_Norm == 'BS 6', 'Emission_Norm'] = 'BS VI'
```

- ### Renaming Column & Values
`Fuel_Tank_Capacity:` The fuel tank capacity is now presented in litres(_and data type is float_) 

```python
cars['Fuel_Tank_Capacity'] = cars.Fuel_Tank_Capacity.str[:-7].astype('float')
```

```python
cars.Fuel_Tank_Capacity.mean(), cars.Fuel_Tank_Capacity.median()
```

Imputing average range for fuel tank for all cars

```python
cars.loc[cars.Fuel_Tank_Capacity.isna(), 'Fuel_Tank_Capacity'] = 55
```

- ### Renaming Column & Values
`Ex-Showroom_Price:` The price represented in Rupees with no formatting(_and data type is int_) 

```python
cars['Ex-Showroom_Price'] = cars['Ex-Showroom_Price'].str[4:].str.replace(',','').astype('int')
```

Excluding null values for height, length, width, body_type columns

```python
cars.dropna(subset=['Height', 'Length', 'Width', 'Body_Type'], inplace=True)
```

```python
cars.loc[cars.City_Mileage == '?23.6 km/litre', 'City_Mileage'] = '23.6 km/litre'
```

- ### Renaming Column & Values
`City_Mileage:` The mileage represented in ___'km/litre'___ with no formatting(_and data type is float_) 

```python
cars.City_Mileage = cars.City_Mileage.str[:-9]
cars.City_Mileage = cars.City_Mileage.str.replace(',','.')
```

```python
cars.loc[cars.City_Mileage == '12.5-12.7', 'City_Mileage'] = '12.6'
cars.loc[cars.City_Mileage == '26032', 'City_Mileage'] = '26.32'
```

```python
cars.City_Mileage = cars.City_Mileage.astype('float')
```

Imputing city mileage to its average value that is around 15

```python
cars.City_Mileage.fillna(round(cars.City_Mileage.mean(), 2), inplace=True)
```

<mark>NOTE: </mark>_Leaving highway mileage column same; will remove later_

- ### Renaming Column & Values
`ARAI_Certified_Mileage:` The mileage represented in ___'km/litre'___ with no formatting(_and data type is float_)

```python
cars.ARAI_Certified_Mileage = cars.ARAI_Certified_Mileage.str[:-9]
cars.loc[cars.ARAI_Certified_Mileage == '9.8-10.0', 'ARAI_Certified_Mileage'] = '9.9'
cars.loc[cars.ARAI_Certified_Mileage == '22.4-21.9', 'ARAI_Certified_Mileage'] = '22.1'
```

```python
cars.ARAI_Certified_Mileage = cars.ARAI_Certified_Mileage.astype('float')
```

Imputing average mileage value for ARAI_Certified_Mileage column

```python
cars.ARAI_Certified_Mileage.fillna(round(cars.ARAI_Certified_Mileage.mean(), 2), inplace=True)
```

<mark>NOTE: </mark>Leaving ARAI_Certified_Mileage column untouched as it contain lot of missing values

- ### Renaming Column & Values
`Kerb_Weight:` The weight represented in ___'kg'___ with no formatting(_and data type is int_)

```python
cars.Kerb_Weight = cars.Kerb_Weight.str[:-3]
cars.loc[cars.Kerb_Weight == '1053-1080', 'Kerb_Weight'] = '1060'
cars.loc[cars.Kerb_Weight == '1016-1043', 'Kerb_Weight'] = '1030'
cars.loc[cars.Kerb_Weight == '1225 kg', 'Kerb_Weight'] = '1225'
cars.Kerb_Weight = cars.Kerb_Weight.astype('Int64')
```

Imputing kerb weight to its average value

```python
cars.Kerb_Weight.fillna(round(cars.Kerb_Weight.mean()), inplace=True)
```

<mark>NOTE: </mark>Leaving Gears column null values as same

```python
cars.loc[cars.Gears == '7 Dual Clutch', 'Gears'] = '7'
cars.Gears = cars.Gears.astype('str')
```

```python
cars.Gears = np.where(cars.Gears == 'nan', np.nan, cars.Gears)
# cars.Gears = np.where(cars.Gears == 'Unknown', np.nan, cars.Gears)
```

```python
cars.Gears.value_counts(dropna=False)
```

```python
cars[cars.Gears.isna()]
```

- ### Renaming Column & Values
`Ground_Clearance:` The ground clearance represented in ___'mm'___ with no formatting(_and data type is float_)

```python
cars.Ground_Clearance = cars.Ground_Clearance.str[:-3].astype('float')
```

```python
print(cars.Ground_Clearance.describe())
```

Imputing ground clearance null values to 50th percentile

```python
cars.Ground_Clearance.fillna(round(cars.Ground_Clearance.median()), inplace=True)
```

___Check the data types of each column. Ensure they are appropriate for the analysis.___

Converting `Displacement` col. dtype

```python
cars.Displacement = cars.Displacement.str[:-3].astype('int')
cars.Displacement
```

Converting `Seating_Capacity` col. dtype

```python
cars.Seating_Capacity = np.where(cars.Seating_Capacity.isna(), 0, cars.Seating_Capacity)
cars.Seating_Capacity = cars.Seating_Capacity.astype('int')
cars.Seating_Capacity = cars.Seating_Capacity.astype('str')
cars.Seating_Capacity = np.where(cars.Seating_Capacity == '0', np.nan, cars.Seating_Capacity)
cars.Seating_Capacity.unique()
```

Converting `Wheelbase` col. dtype

```python
cars.Wheelbase = cars.Wheelbase.str[:-3]
cars.dropna(subset='Wheelbase', inplace=True)
cars.Wheelbase = pd.to_numeric(cars.Wheelbase, downcast='integer')
```
